# Dataset RB circuits

```python
import matplotlib.pyplot as plt
import seaborn as sns

from qiskit.providers.fake_provider import FakeLima

from blackwater.data.generators.rb import rb_generator
```

```python
lima = FakeLima()
gen = rb_generator(
    max_iter=100, backend=lima, n_qubits=5, num_samples=1, lengths=[1]
)
```

```python
entry, circuit, observable = next(gen)

print(f"Observable: {observable}")
print(f"Entry: {entry}")
print("Circuit:")
circuit.draw(fold=-1)
```

```python
ideal = []
noisy = []
for entry, circuit, observable in gen:
    ideal.append(entry.ideal_exp_value)
    noisy.append(entry.noisy_exp_value)
```

```python
sns.histplot((ideal, noisy), kde=True)
plt.title("generated RB circuits exp values")
plt.xlabel("Exp value")
plt.show()
```
